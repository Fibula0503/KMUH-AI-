# 數值資料標準化

## 對應字典法

```python
import pandas as pd
df = pd.read_csv('客戶聯絡狀況資料檔.csv')

df['婚姻'].unique()
```

```python
dict1 = {'單身':1, '已婚':2, '離婚':3, '未知':4}
df['婚姻'].replace(dict1, inplace=True)
df
```

```python
dict2 = {1:'單身', 2:'已婚', 3:'離婚', 4:'未知'}
df['婚姻'].replace(dict2, inplace=True)
df
```

```python
df['學歷'].unique()
```

```python
dict3 = {'高中':1, '專科':2, '大學':3, '國中':4, '小學':5, '未知':6, '文盲':7}
df['學歷'] = df['學歷'].map(dict3)
df
```

# 使用Pandas 進行特徵選擇

## 皮爾森(Pearson) 相關係數

```python
import pandas as pd
df = pd.read_csv('地區房價資料檔.csv')
featuresCorr = df.corr()
featuresCorr
```

```python
targetCorr = featuresCorr['房價']
targetCorr = targetCorr.drop('房價')
selectedFeatures = targetCorr[abs(targetCorr) > 0.4]
print("選擇特徵數： {} \n選擇特徵:\n{}".
      format(len(selectedFeatures), selectedFeatures))
```

## 肯德爾(Kendall) 相關係數

```python
import pandas as pd
df = pd.read_csv('地區房價資料檔.csv')
featuresCorr = df.corr('kendall')
featuresCorr
```

```python
targetCorr = featuresCorr['房價']
targetCorr = targetCorr.drop('房價')
selectedFeatures = targetCorr[abs(targetCorr) > 0.4]
print("選擇特徵數： {} \n選擇特徵:\n{}".
      format(len(selectedFeatures), selectedFeatures))
```
